## Series from list


```python
import numpy as np 
import pandas as pd
```

```python
# lets create a series in string
country = ["india","nepal","sirlanka","bhutan","bangaladesh"]
pd.Series(country)
# from pandas.Series we get the same list as a series where  i get a indexing and values names along with it 
```

```python
# lets create a series in integer
runs = [10,45,78,100,274]
pd.Series(runs)
```

```python
# lets create a user defined indexing for the series 
marks = [90,98,97,99,100] # this is the marks of the some subjects
subjects = ['math','science','hindi','sst','english']

# now we want ki ham subject ko as a index le and marks ko as a vlaues
pd.Series(marks, index=subjects) # ham explicitlly subejcts ko index bana rahe hai 
```

```python
# we can store the all this in a variables
mark =pd.Series(marks, index=subjects, name='Ujjwal ke marks') # we can store marks in it also and print it 
mark # after writting the logic in an varibale we can print it also just wile writing the name
# print(mark)
```

## Series By using dict

```python
# lets make the series while using the dictionary
marks = {
    'science' : 100,
    'english' : 90,
    'sst' : 98,
    'hindi' : 90,
    'math' : 100
} # here we create same data as the above

marks_series = pd.Series(marks, name="Ujjwal ke marks")
marks_series 
```
