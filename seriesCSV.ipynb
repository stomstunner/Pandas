## Series with read_csv

```python
import numpy as np
import pandas as pd 
```

### with one coloumn

```python
# importing the subs.csv // comma seperated values 
pd.read_csv('subs.csv') # here we give the path of the file 

# in pandas we have two types of data 
# series and
# pandas
# toh normally the mthod read_csv store the data ina dataframe
# if want explicitly ki ham data ko change kare series me then hame file path ke badd comma deke squeeze= True dena hoga

# pd.read_csv('subs.csv', squeeze= 1) ab ye parameter me dena hat gaya

data_csv = pd.read_csv('subs.csv')
data_csv.squeeze  #isse ham dataframe ko series me convert kar liya /
# we can also see ki ham table ka naam bhi dekh sakte hai
```

```python
# here we see ki ham 2 coloum  wale csv file ko kaisse read karte hai
# because we have to use the squeeze method so we have ti store it in the varibales
data_kholi = pd.read_csv('kohli_ipl.csv', index_col='match_no')['runs'] # we can make the one coloumn in the series , just we have to write it in the array
# just like the above 
data_kholi.squeeze

```

```python
# now we load the third file
films = pd.read_csv("bollywood.csv",index_col="movie")
films.squeeze # if we do not write squeeze then the read_csv make its a dataframe
# and if we write the squeeze then the data frame becomes series
```

## head and tail method

```python
# head
# when we want preview of top 5 elements or custom top elements
films.head()
films.head(7) # we can pass the number of elements we want

# tail
# when we want ki ham last ke 5 elements preview kare ya custom
films.tail()
films.tail(10) # by passing values we can see only 60 values only 
```

## Sample Mthod

```python
# sample gives us the random 1 row from the table and it is comletly random like har baar same code ko chalane ke koi dusra hi row deta hai
films.sample()
# we can file custom sample file also by just passing ki hame kitna sample row chaiye
films.sample(4)
```

## value_count Method

```python
# value_count() gives the frequency of a vlaues in the table
# like koi vlaue kitni baar table me aa rha hai
films.value_counts() # in the decending order
```

## sort_vlaues 

```python
# here we see how we can find the values in sorted order
data_kholi.sort_values() # by defalut ye hame ek acending order deta hau with the vlaues

# if we want ki ham decending me sort kare tho
data_kholi.sort_values(ascending=False)

# now if we want ki ham top ka vlaue hi dekhe
data_kholi.sort_values(ascending=False).head(1)

# now if we want ki ham bass runs ko hi dekhe
data_kholi.sort_values(ascending=False).head(1).values

# now if we want ki ham bass ek number ko hi dekhe toh na ki array ko
data_kholi.sort_values(ascending=False).head(1).values[0].item()
```

```python
# we can change the original table with the help of sort_vlaues -> ander -> inplace=True isse ham sirf ek preview nahi dekhte balki original vlaues ko change kar skate hai

data_kholi.sort_values(inplace=True) # isse kuch output nahi aayega baliki original vlaues change ho jayega

data_kholi
```

## sort_index() mathod is used to sort the index

```python
# let sort the movies ka index -> means movies ka name with the accending order to the alphabets
films.sort_index() # ye hamsha hame  digit -> a -> z deta hai for strings

# we can use inplace for fixing the same sorting in the main table
films.sort_index(inplace=True)
films
```
